# Heart Disease Prediction - Inference Demo

This notebook demonstrates how to use the trained model for inference.

````python
import sys
import os
sys.path.insert(0, os.path.abspath('..'))

import pandas as pd
import numpy as np
import joblib
from src.predict import HeartDiseasePredictor, create_sample_input, FEATURE_NAMES, FEATURE_DESCRIPTIONS

print('Libraries imported!')
````

````python
# Load the predictor
predictor = HeartDiseasePredictor(
    model_path='../models/model.pkl',
    preprocessor_path='../models/preprocessor.pkl'
)
print('Model and preprocessor loaded!')
````

````python
# Feature descriptions
print('Input Features:')
print('=' * 60)
for feature, desc in FEATURE_DESCRIPTIONS.items():
    print(f'{feature:12} : {desc}')
````

````python
# Sample prediction
sample = create_sample_input()
print('Sample Input:')
print(sample)

result = predictor.predict(sample)
print('\nPrediction Result:')
print(result)
````

````python
# Multiple predictions
test_cases = [
    {'age': 45, 'sex': 1, 'cp': 3, 'trestbps': 120, 'chol': 200, 'fbs': 0, 'restecg': 0, 'thalach': 170, 'exang': 0, 'oldpeak': 0.5, 'slope': 1, 'ca': 0, 'thal': 3},
    {'age': 65, 'sex': 1, 'cp': 4, 'trestbps': 160, 'chol': 280, 'fbs': 1, 'restecg': 2, 'thalach': 120, 'exang': 1, 'oldpeak': 3.0, 'slope': 3, 'ca': 2, 'thal': 7},
    {'age': 55, 'sex': 0, 'cp': 2, 'trestbps': 140, 'chol': 240, 'fbs': 0, 'restecg': 1, 'thalach': 150, 'exang': 0, 'oldpeak': 1.5, 'slope': 2, 'ca': 1, 'thal': 6},
]

print('Batch Predictions:')
print('=' * 80)
for i, case in enumerate(test_cases, 1):
    result = predictor.predict(case)
    print(f'\nCase {i}: {result["prediction_label"]} (Confidence: {result["confidence"]:.2%}, Risk: {result["risk_level"]})')
````
